```python
page_num =1
alpha = 'a'
params = {
    'mode':'BOTH',
    'page':page_num,
    'records':40,
    'filterActive':'false',
    'filterTeamId':'6',
    'filterAlpha': alpha,
    'filterFormatLevel': 'INTERNATIONAL',
    'sort': 'ALPHA_ASC'
}
teams_base_api = 'https://hs-consumer-api.espncricinfo.com/v1/pages/player/search'
```

```python
response = requests.get(teams_base_api,params=params)
data = json.loads(response.content)
```

```python
data
```

```python
class PlayerUrl:
    def __init__(self,slug_name,id):
        self.id = id
        self.slug_name = slug_name
    def get_url(self):
        return f"{self.slug_name}-{self.id}"
```

```python
players_slug_data = []
```

```python
for player in data['results']:
    slug_name = player['slug']
    id = player['objectId']
    players_slug_data.append(PlayerUrl(slug_name=slug_name,id=id))
```

```python
for player in players_slug_data:
    print(player.get_url())
```

```python
from datetime import datetime
dt = datetime(year = 1996,month = 3, day = 23)
```

```python
def parse_date_from_json(date):
    if not date:
        return None
    year = date.get('year')
    month = date.get('month')
    day = date.get('date')
    return datetime(year= year, month = month , day = day)

class Player:
    def __init__(
        self,
        slug = None,
        id = None,
        long_name = None,
        gender = None,
        image_url = None,
        headshot_image_url=None,
        dob=None,
        dod=None,
        country_team_id=None
    ):
        self.slug = slug
        self.id = id
        self.long_name = long_name
        self.gender = gender
        self.image_url =  image_url
        self.headshot_image_url = headshot_image_url
        self.dob = dob
        self.dod = dod
        self.country_team_id = country_team_id
  
    
    def get_player_url(self):
        if self.slug and self.id:
            return f"{self.slug}-{self.id}"
        return None
    
    
    def player_from_json(player):
        headshot_image = player.get('headshotImage')
        if headshot_image:
            headshot_image_url = headshot_image.get('url')
        else:
            headshot_image_url=None
        return Player(
            slug= player.get('slug'),
            id=player.get('objectId'),
            long_name=player.get('longName'),
            gender = player.get('gender'),
            image_url= player.get('imageUrl'),
            headshot_image_url=headshot_image_url,
            dob= parse_date_from_json(player.get('dateOfBirth')),
            dod = parse_date_from_json(player.get('dateOfDeath')),
            country_team_id=player.get('countryTeamId')
            
        )
```

```python
players_list = []
```

```python
for player in data['results']:
    players_list.append(Player.player_from_json(player))
```
